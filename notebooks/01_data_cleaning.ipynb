```python
import pandas as pd

# Load dataset
df = pd.read_csv(
    "../data/raw/Global_Superstore2.csv",
    encoding="latin1"
)

df.head()
```

```python
print("Shape:", df.shape)
print("\nData Types:\n")
print(df.info())

print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
# Check duplicates
print("Duplicate rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

print("Shape after removing duplicates:", df.shape)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df.info()
```

```python
df["Unit_Price"] = df["Sales"] / df["Quantity"]
```

```python
df[["Sales", "Quantity", "Unit_Price"]].head()
```

```python
# Remove zero or negative quantity
df = df[df["Quantity"] > 0]

# Remove zero or negative sales
df = df[df["Sales"] > 0]

# Remove zero or negative price
df = df[df["Unit_Price"] > 0]

print("Shape after removing invalid values:", df.shape)
```

```python
df.to_csv("../data/processed/cleaned_superstore.csv", index=False)
```
