```python
# remove duplicates
df.drop_duplicates(inplace=True)

# fill numeric columns with mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# fill categorical columns with mode
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
import pandas as pd

df = pd.read_csv("retail_store_inventory.csv")

# convert to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.month_name()
```

```python
yearly_data = df.groupby('Year').sum(numeric_only=True)
print(yearly_data)
```

```python
monthly_data = df.groupby('Month_Name').sum(numeric_only=True)
print(monthly_data)
```

```python
year_month = df.groupby(['Year','Month']).sum(numeric_only=True)
print(year_month)
```

```python
df['YearMonth'] = df['Date'].dt.to_period('M')
grouped = df.groupby('YearMonth').sum(numeric_only=True)
print(grouped)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns)
```
